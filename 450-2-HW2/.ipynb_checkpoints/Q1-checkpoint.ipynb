## value distribution

```python
# Plot the figure
fig=plt.subplots(figsize=(10,5))
fig_cdf = plt.subplot( 1,2,1)
fig_pdf = plt.subplot( 1,2,2)
private_values = np.arange(0,5,0.2)

# --- 1. true distribution
pdf_true = stats.lognorm.pdf(private_values, 
                             s= auction.dist_par_sigma, loc = 0, scale = np.exp(auction.dist_par_mu))
cdf_true = stats.lognorm.cdf(private_values, 
                            s= auction.dist_par_sigma, loc = 0, scale = np.exp(auction.dist_par_mu))
fig_pdf.plot(private_values, pdf_true, label='true pdf '.format(n) )
fig_cdf.plot(private_values, cdf_true, label='true cdf '.format(n) )

# ---- 2.estimated distribution
for n in range(2,6):
    
    # A. simulation
    (_, all_bids, _) = auction.simulate_auctions(n_auctions = 1000, n_bidders=n)
    
    # B. stimation
    fpa_est = FPA.estimate_FPA( n_bidders=n, all_bids = all_bids )
    fpa_est.estimate_private_value_distribution()

    # C. get values
    pdfs = fpa_est.pv_dist.pdf( private_values)
    cdfs = fpa_est.pv_dist.cdf( private_values)
    
    # D. plot
    fig_pdf.plot(private_values, pdfs, label='N={} pdf'.format(n) )
    fig_cdf.plot(private_values, cdfs, label='N={} cdf'.format(n) )

fig_pdf.legend(loc='best');
fig_cdf.legend(loc='best');
```
