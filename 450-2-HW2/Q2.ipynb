## (c)

Below we plot the simulation results from different sigma values. Several take aways:
1. The larger the sigma is, the less frequent we get a new draw.
   As we could see from the process, there are several "straight lines" when sigma=10.
   This is because the updating probability is punishing it by setting a lower probability when x_tilde is far from x. 
2. The smaller the sigma is, the more frequent we get a new draw, but the new draw would be super closed to the old one.
   This means we are not able to touch the region that is far from the initial value.
   
Therefore, the choice of sigma should balance the frequency of changing and the distance of each change. 

1. If the frequency is too low, then the process would get stuck at random value where the true distribution does not have a high density. 
2. If the frequency is high, the sacrifice is that the distance of each change is low. If each step is too low, then the process get stuck at the neighborhood and cause the same problem.

That's why when sigma = 0.1, the simulated process is able to characterize one part of the mixure normal very well but not able to touch the other part of the mixture normal.

When sigma = 3 or 10, although the process is able to reach all possible region so we get a roughly mixture look, it can not characterize the distrbution very good locally. Locally the dense would be at the random places it ever went to. Therefore it looks like a mixture of something but not the true normal distributions.

## (e)

Ideally, we want the simulation of the true distribution to be iid draws. However, the approach we are taking are path dependent. When we have to chose a q s.t. the frequency of changing is low (e.g. when sigma is high in the above setting), the random draws are not independent. And eventhough the frequency of changing is high, x_tilde is dependent on x by definition of q. (The dependence is higher when sigma is low.) 

Therefore, we can pick every Kth draw to avoid the dependence.
